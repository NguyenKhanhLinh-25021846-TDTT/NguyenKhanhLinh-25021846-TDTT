```python
#Exercise 1: Nhập vào 3 số nguyên dương a,b,c. Viết chương trình kiểm tra xem đây là tam giác đều, tam giác cân, tam giác vuông hay tam giác thường
a,b,c = map(int,input("Nhap 3 so nguyen duong: ").split())
if a > 0 and b > 0 and c > 0 and a + b > c and b + c > a and c + a > b:
    if a == b == c:
        print("Tam giac deu")
    elif a**2 + b**2 == c**2 or b**2 + c**2 == a**2 or c**2 + a**2 == b**2:
        print("Tam giac vuong")
    elif a == b or b == c or c == a:
        print("Tam giac thuong")
    else: print("Tam giac thuong")
else: print("Khong phai tam giac")
```

```python
# Exercise 2: Nhập vào số nguyên dương n. Viết chương trình in ra số cách phân tích của n thành các thừa số nguyên tố
import math
n = int(input())
output = ""
def so_nguyen_to(a):
    if a == 1: return False
    for i in range (2,math.isqrt(a) + 1):
        if a % i == 0:
            return False
    else: return True
while n > 0:
    for j in range(2,n//2 + 1):
        if so_nguyen_to(j):
            if n % j == 0:
                output = f"{output}*{j}"
                n = n//j
                continue
print(output)
```

```python
#Dãy số fibonacci: 1, 1, 2, 3, 5,... có số kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào A, tìm số trong dãy fibonacci lớn nhất nhưng không > A
A = int(input())
if A < 1:
    print("Khong so nao phu hop")
else: 
    a = 1
    b = 1
    while b <= A:
        a,b = b,a + b
print(a)
```

```python
#Hàm chữ số khác nhau
def different(a):
    cso1 = 0
    while a > 0:
        cso = a % 10
        a //= 10
        if cso != cso1:
            cso1 = cso
        else: return False
    return True
```

```python
#Đẳng cấu
def dang_cau(a: str,b: str) -> bool:
    dict = {}
    used = set()
    for i in range(len(a)):
        x = a[i]
        y = b[i]
        if x in dict:
            if dict[x] != y: return False
        else:
            if y in used: return False
        dict.[x] = y
        used.append(y)
    return True
```

```python
#Cho dãy số nguyên. Hãy tìm phần tử có số lần xuất hiện nhiều nhất trong dãy. Nếu có nhiều phần tử cùng tần suất, chọn phần tử xuất hiện trước trong danh sách
n = list(map(int,input().split()))
socantim = 0
max_count = 1
m = []
for i in range(len(n)):
    if n[i] not in m:
        solan = 1
        for j in range(i + 1, len(n)):
            if n[i] == n[j]:
                solan += 1
        if solan > max_count:
            max_count = solan
            socantim = n[i]
    m.append(n[i])
print(socantim)
```

```python
#Gọi S(n) là tổng các ước đúng của n. Một số nguyên dương n được phân loại như sau:...
import math
n = int(input())
tong = 0
for i in range(1,math.isqrt(n) + 1):
    if n % i == 0:
        tong += i
        if i != 1 and i*i != n:
            tong += n//i
if tong == n: print("0")
elif tong < n: print("-1")
elif tong > n: print("1")
```

```python
#Nhập 1 xâu từ bàn phím và đếm số lượng chữ số trông xâu
n = input()
count = 0
for ch in n:
    if ch.isdigit():
        count += 1
print(count)
```

```python
#Nhập một xâu từ bàn phím và tính tổng chữ số trong xâu. Tổng các chữ số hoặc -1 nếu xâu không chứa chữ số
n = input()
sum = 0
co_chu_so = False
for ch in n:
    if ch.isdigit():
        sum += int(ch)
        co_chu_so = True
if co_chu_so:
    print(sum)
else: print("-1")
```

```python
#Cho một số nguyên không âm n do người dùng nhập từ bàn phím. Hãy viết chương trình Python để tính số Fibonacci thứ n 
n = int(input())
a = 1
b = 1
if n == 0:
    print("0")
elif n == 1 or n == 2:
    print("1")
else:
    p = 2
    while p != n:
        a,b = b,a + b
        p += 1
    print(b)
```
