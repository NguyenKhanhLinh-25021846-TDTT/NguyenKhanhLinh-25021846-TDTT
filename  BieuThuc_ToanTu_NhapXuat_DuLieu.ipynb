```python
#W2A3
a = int(input())
b = int(input())
print("Tổng:",a + b)
print("Hiệu:",a - b)
print("Tích:",a * b)
if b != 0 :
    print("Chia lấy phần nguyên:",a // b)
    print("Chia lấy phần dư:",a % b)
    print("Chia lấy 2 chữ số thập phân:", f"{a / b:.2f}")    # :.2f → định dạng số thực với 2 chữ số sau dấu thập phân
else:
    print("Không thể chia cho 0")
```

```python
#W2A4
a1, b1, c1, a2, b2, a3 = map(int,input().split())
# map(int,...) biến chuỗi thành số nguyên
# split() cho phép nhập nhiều số cách nhau bởi dấu cách
TB = ((a1+b1+c1)+(a2+b2)*2+a3*3)/10
print("Điểm trung bình môn học là", f"{TB:.1f}")
```
