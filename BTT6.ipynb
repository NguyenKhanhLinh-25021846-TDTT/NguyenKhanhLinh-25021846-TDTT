```python
#W6A2
n = list(map(int,input("Nhap cac so nguyen: ").split()))
m = []
for i in range(len(n)):
    sum = 0
    for j in range(0,i + 1):
        sum += n[j]
    m.append(sum)
print(m)
```

```python
#W6A4
a = input()
pairs = a.split()
d = {}
for i in pairs:
    key,value = i.split(":")
    if key not in d:
        d[key] = []
    d[key].append(value)
print(d)
```

```python
#W6A4
a = input()
pairs = a.split()
d = {}
for i in pairs:
    key,value = i.split(":")
    d.setdefault(key,[]).append(value) #dict.setdefault(key,default) nếu key tồn tại thì giữ nguyên value, không tồn tại thì bằng default 
print(d)
```

```python
#W6A5
a = list(map(int,input().split()))
d = {'positives': 0, 'negatives': 0, 'zeros': 0}
for i in a:
    if i > 0: d['positives'] += 1
    elif i < 0: d['negatives'] += 1
    elif i == 0: d['zeros'] += 1
print(d)
```

```python
#W629
a = input("Nhập các cặp key:value của danh sách 1: ")
b = input("Nhập các cặp key:value của danh sách 2: ")
pairs1 = a.split()
pairs2 = b.split()
d1 = {}
for i in pairs1:
    k1,v1 = i.split(":")
    d1[k1] = [v1]
for j in pairs2:
    k2,v2 = j.split(":")
    d1.setdefault(k2,[]).append(v2)
print(d1)
```

```python
#W6A10
a = list(map(int,input().split()))
k = int(input())
a.sort()
lst = []
for i in range(0,len(a)):
    for j in range(i + 1,len(a)):
        if a[i] + a[j] == k:
            if (a[i],a[j]) not in lst:
                lst.append((a[i],a[j]))
print(lst)
```

```python
#W6A12
lst = list(map(int,input().split()))
lst.sort()
num = lst[0]
m = 0
cnt = 0
for i in range(0,len(lst)):
    if lst[i] != lst[i - 1]:
        if cnt > m: 
            m = cnt
            num = lst[i - 1]
        cnt = 1
    else: cnt += 1
print(num)
```
