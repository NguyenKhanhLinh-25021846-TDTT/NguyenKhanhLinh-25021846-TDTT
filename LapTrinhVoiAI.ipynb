```python
#W1A1
a = 7        # tạo biến a và gán cho nó giá trị 7(int)
b = 5        # tạo biến b và gán cho nó giá trị 5(int)
c = a - b    # thực hiện phép 7-5=2 và gán cho biến c giá trị 2
print(c)     # in giá trị c ra màn hình
# Kết Quả(2) vì c = 2
```

```python
#W1A2
city = "Hà Nội"    # tạo và gán cho biến city chuỗi "Hà Nội"(str)
year = 2025        # tạo và gán cho biến year giá trị 2025(int)
print("Thành phố: ", city, " - Năm: ", year)    # in các đối số ra theo thứ tự
# Kết Quả(Thành phố:  Hà Nội  - Năm:  2025) vì Python sẽ tự động chèn một dấu cách giữa các đối số khi in ra
```

```python
#W1A3
numbers = [1, 2, 3, 4]    # Tạo 1 danh sách tên numbers(list) chứa các số nguyên 1, 2, 3, 4(int).
for x in numbers:         # Vòng lặp for lặp qua từng phần tử trong numbers .Dấu ':' và thụt lề chỉ ra khối lệnh thuộc về vòng lặp
    if x % 2 == 0:        # Nếu số dư của phép x chia 2 bằng 0 
        print(x, "là số chẵn")    # in ra giá trị x và chuổi "là số chẵn"
    else:                 # Nếu điều kiện của if sai hay x không chia hết cho 2 
        print(x, "là số lẻ")      # in ra giá trị x và chuổi "là số lẻ"
```

Kết Quả(
1 là số lẻ
2 là số chẵn
3 là số lẻ
4 là số chẵn)
vì 
Khi x = 1: 1 % 2 = 1 ≠ 0 → in 1 là số lẻ.

Khi x = 2: 2 % 2 = 0 → in 2 là số chẵn.

Khi x = 3: 3 % 2 = 1 → in 3 là số lẻ.

Khi x = 4: 4 % 2 = 0 → in 4 là số chẵn.

```python
#W1A5
animals = ["cat", "dog", "cat", "bird"]   # Tạo danh sách gồm 4 phần tử
count = 0                                 # Biến đếm, ban đầu = 0
for a in animals:                         # Lặp qua từng phần tử trong danh sách
    count += 1                            # Mỗi lần lặp, tăng count thêm 1
print("Số phần tử trong danh sách là:", count)  # In chuỗi "Số phần tử trong danh sách là:" và kết quả count cuối cùng
# Kết Quả(Số phần tử trong danh sách là: 4) vì list animals có 4 phần tử nên sau vòng lặp count = 4
```

```python
#W1A8
for i in range(3):                  # vòng lặp chạy 3 lần, i = 0, 1, 2
    print("AI đang học lần", i + 1) # Vì i chạy từ 0 → 2 nên i + 1 chạy từ 1 → 3
print("Huấn luyện xong!")           # nằm ngoài vòng lặp, chỉ in 1 lần
```
