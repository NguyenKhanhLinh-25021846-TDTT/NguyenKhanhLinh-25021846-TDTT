```python
#W4A1_1
try:
    n = int(input("Nhập số số nguyên nhỏ hơn hoặc bằng 1000:"))
    tong = 0
    if n > 0:
        if n <= 1000:
            for x in range(1,n+1):
                tong +=x
            print(tong)
        else:
            print("Số đã nhập không nhỏ hơn hoặc bằng 1000")
    else:
        print("Hình như bạn đã nhập số nguyên âm")
except ValueError:
    print("Đây không phải số nguyên")
```

```python
#W4A8
n = int(input("Nhập số nguyên dương: "))
def reverse(n):
    n_rv = 0
    while n > 0:
        n_rv = n_rv*10 + n % 10
        n = n // 10
    return n_rv
buoc = 0
while n != reverse(n):
    n += reverse(n)
    buoc += 1
print(buoc,n)
```

```python
#W4A9
import math
n = int(input("Nhập số nguyên dương: "))
def chinh_phuong(x):
    if math.sqrt(x) % 1 == 0:
        return True
def cso_khac(y):
    while y > 0:
        cso1 = y % 10
        y = (y-cso1)/10
        cso2 = y % 10
        if cso2 != cso1:
            continue
        else: 
            return False
    else: return True
for i in range(1,n + 1):
    if chinh_phuong(i) and cso_khac(i):
        print(i,end =' ')
```

```python
#W4A10
n = int(input("Nhập số nguyên dương: "))
max_dai = 1
so = 1
for i in range(1,n + 1):
    dai = 1
    x = i
    while x != 1:
        if x % 2 == 0:
            x = x/2
        else: x = x*3 + 1
        dai += 1
    if dai > max_dai:
        max_dai = dai
        so = i
print(so, max_dai)
```
