```python
#W5A4
def so_hoan_hao(n):
    tong = 0
    for i in range((n+1)/2):
        if n % i == 0:
            tong += i
    if n == tong:
        return"Số hoàn hảo"
    else:
        return"Không là số hoàn hảo"
```

```python
#W5A7
def may_tinh(num1: int,pheptoan: str,num2: int):
    if pheptoan == '+':
        return num1 + num2
    elif pheptoan == '-':
        return num1 - num2
    elif pheptoan == '/':
        return num1/num2
    elif pheptoan == '*':
        return num1*num2
print(may_tinh(6,'-',5))
```

Thuật toán đổi số sang dãy bits

1.Lấy số n
2.Chia n cho 2
3.Ghi lại phần dư (0 hoặc 1)
4.Lấy thương chia tiếp cho 2
5.Lặp lại cho đến khi thương bằng 0
->Dãy bit cuối cùng là dãy các phần dư ghi theo thứ tự ngược lại

```python
#W5A8
def nhi_phan(n: int):
    day_bit = 0
    hang = 1
    while n > 0:
        x = n % 2
        n = n // 2
        day_bit = day_bit + x*hang
        hang = hang*10
    return day_bit
def Hamming(a: int,b: int):
    s = 0
    while a > 0:
        cso1 = a % 10
        a = (a - cso1)//10
        while b > 0:
            cso2 = b % 10
            b = (b - cso2)//10
            if cso2 == cso1: pass
            else: s +=1
    return s
a, b = map(int,input("Nhập hai dãy bits: ").split())
print(Hamming(nhi_phan(a),nhi_phan(b)))
```

```python
#W5A9
def tong(x):
    tong = 0
    while x > 0:
        cso = x % 10
        x = (x - cso)//10
        tong += cso
    return tong
n = int(input("Nhập số nguyên dương: "))
print(tong(n))
```
