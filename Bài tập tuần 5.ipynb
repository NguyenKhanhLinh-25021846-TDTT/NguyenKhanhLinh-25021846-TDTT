```python
def Max_Of_Two(a,b):
    if isinstance(a, int) and isinstance(b, int):
        if a > b:
            return a
        elif a < b:
            return b
        else:
            return("Hai số bằng nhau")
    else:
        return"Không phải số nguyên"
print(Max_Of_Two(7, 9.0))
```

```python
#W5A2
def swap(a,b):
    if isinstance(a,int) and isinstance(b,int):
        return(b, a)
    else:
        return"Không phải số nguyên"
print(swap(4,5))
```

```python
#W5A3
def so_nguyen_to(n):
    if isinstance(n,int):
        for i in range(1,(n+1)/2):
            if n % i == 0:
                return"Không phải số nguyên tố"
        return"Là số nguyên tố"
    else:
        return"Không phải số nguyên"
```

```python
#W5A4
def so_hoan_hao(n):
    tong = 0
    for i in range((n+1)/2):
        if n % i == 0:
            tong += i
    if n == tong:
        return"Số hoàn hảo"
    else:
        return"Không là số hoàn hảo"
```

```python
#W5A5
def ton_tai_gia_tri(s,k):
    vi_tri = 1
    for x in s:
        if x == k:
            break
        else:
            vi_tri += 1
    if vi_tri <= len(s):
        return vi_tri
    else:
        return"-1"
```

```python
#W5A6
def giai_thua(n):
    giaithua = 1
    for i in range(1,n + 1):
        giaithua = giaithua*i
    return giaithua
```
