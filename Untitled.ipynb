```python
import math
def xap_xi (x: float,y: float):
    return abs(x - y) < 10**(-6)
a,b,c = map(float,input().split())
if a > 0 and b > 0 and c > 0 and a + b > c and b + c > a and c + a > b:
    if xap_xi(a,b) and xap_xi(b,c):
        print("Tam giác đều")
    elif xap_xi(a**2 + b**2,c**2) or xap_xi(b**2 + c**2,a**2) or xap_xi(c**2 + a**2,b**2):
        print("Tam giác vuông")
    elif xap_xi(a,b) or xap_xi(b,c) or xap_xi(c,a):
        print("Tam giác cân")
    else:
        print("Tam giác thường")
    P = a + b + c
    p = P/2
    S = math.sqrt(p*(p - a)*(p - b)*(p - c))
    print(f"Chu vi: {P:.2f}. Diện tích: {S:.2f}")
else: 
    print("Khong la ba canh cua mot tam giac")
```
