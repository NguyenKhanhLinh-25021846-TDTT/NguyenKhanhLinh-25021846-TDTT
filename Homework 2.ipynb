```python
#Bài 3
import math
a,b,c = map(float,input().split())
if a + b < c or b + c < a or  c + a < b:
    print("Không là ba cạnh một tam giác")
else: 
    if a == b or b == c or c == a: x = True 
    else: x = False 
    if a**2 + b**2 == c**2 or b**2 + c**2 == a**2 or c**2 + a**2 == b**2:
        y = True
    else:
        y = False
    if x and y: print("Tam giác vuông cân")
    elif x: print("Tam giác đều")
    elif y: print("Tam giác vuông")
    elif a == b == c: print("Tam giác đều")
    else: print("Tam giác thường")
    p = (a + b + c)/2
    S = math.sqrt(p*(p - a)*(p - b)*(p - c))
    print(f"Diện tích là {S:.2f}. Chu vi là {2*p:.2f}")
```

```python
#Bài 4
a,b = map(float,input().split())
if a == b == 0:
    print("Vô số nghiệm")
elif a == 0 != b:
    print("Vô nghiệm")
else: print(f"Nghiệm là {-b/a:.2f}")
```

```python
#Bài 5
import math
a,b,c = map(float,input().split())
delta = b**2 - 4*a*c
if a == 0:
    if b == 0:
        if c == 0:
            print("Vô số nghiệm")
        else: print("Vô nghiệm")
    else: print(f"Nghiệm là {-c/b:.2f}")
else:
    delta = b**2 - 4*a*c
    if delta < 0:
        print("Vô nghiệm")
    elif delta == 0:
        print(f"Nghiệm kép là {-b/(2*a):.2f}")
    elif delta > 0:
        print(f"Hai nghiệm là {(-b - math.sqrt(delta))/(2*a):.2f} và {(-b + math.sqrt(delta))/(2*a):.2f}")
```

```python
#Bài 8
a,b,m,c,d,n = map(float,input().split())
if c*b - a*d == c*m -a*n == 0:
    print("Vô số nghiệm")
elif c*m - a*n == 0 != c*b - a*d:
    print("Vô nghiệm")
else:
    print(f"Nghiệm của phương trình là x = {(b*n - d*m)/(c*b - a*d):.2f} và y = {(c*m - a*n)/(c*b - a*d):.2f}")
```

```python
#Bài 10
s = input("Nhập tọa độ tâm dạng (x,y): ")
r = float(input("Bán kính: "))
A = input("Nhập tọa độ điểm A: ")
s = s[1:-1]
a,b = map(int,s.split(','))
A = A[1:-1]
x,y = map(int,A.split(','))
if (x - a)**2 + (y - b)**2 == r**2:
    print("A thuộc đường tròn")
else: print("A không thuộc đường tròn")
```
