```python
#Bài 3:
c = input("Nhập một kí tự:")
if(len(c)==1):
    if(c.isupper):    #lược bỏ = True
        c_lower = chr(ord(c)+32)
        print(c_lower)
    elif(c.islower):
        c_upper = chr(ord(c)-32)
        print(c_upper)
    else:
        print("Đây không phải là một kí tự Alphabet")
else:
    print("Đây không phải một kí tự duy nhất")
```

```python
#Bài 6:
import math
a, b, c = map(float, input("Nhập vào 3 cạnh của tam giác:").split())
if(a + b > c and a + c > b and b + c > a):
    p = 0.5*(a + b + c)
    S = p*(p-a)*(p-b)*(p-c)
    S = math.sqrt(S)
    print("Diện tích tam giác:",f"{S:.1f}")
else:
    print("Không phải 3 cạnh của tam giác")
```

```python
#Bài 2
a = float(input("Chiều rộng hình chữ nhật: "))
b = float(input("Chiều dài hình chữ nhật: "))
pi = 3.14
if a,b > 0:
  hcn = a*b
  ht = pi*((a/2)**2)
  trong_cay = hcn - ht
  print(f"Diện tích trồng cây: {trong_cay:.2f}")
else:
  print("Kích thước không hợp lệ")
```
