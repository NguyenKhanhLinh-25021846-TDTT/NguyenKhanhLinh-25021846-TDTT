```python
#Bài 14
n = int(input("Nhập một số nguyên: "))
max = n
min = n
while True:
    if n > max: max = n
    if n < min: min = n
    if n != -1:
        n = int(input())
    else: break
print(f"Số lớn nhất là {max} và số nhỏ nhất là {min}")
```

```python
#Bài 19
A = int(input("Nhập số nguyên dương: "))
st = 1
ss = 1
while ss <= A:
    so_can_tim = ss
    a = st
    b = ss
    st = b
    ss = a + b
print(so_can_tim)
```

```python
#Bài 20
n = int(input("Nhập số nguyên dương: "))
def so_nguyen_to(x):
    if x == 1: return False
    for i in range(2,x//2 + 1):
        if x % i == 0:
            return False
    else: return True
count = 0
while n > 0:
    cso = n % 10
    n = (n - cso)//10
    if so_nguyen_to(cso):
        count += 1
print(count)
```
